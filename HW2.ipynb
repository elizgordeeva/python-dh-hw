```python
# -*- coding: utf-8 -*-
from os import walk


def work():
    poems = open("poems.txt", "a")
    path = "poems"
    dict = {}

    for (dirpath, dirnames, filenames) in walk(path):
        print("Current folder %s and list of files %s" % (dirpath, str(filenames)))
        dict[dirpath] = filenames
        print("In folder %s we have %d file(s)" % (dirpath, len(dict[dirpath])))

    for key in dict:
        for current in dict[key]:
            file = open( key + "/" + current, "r", encoding="utf-8")
            text = file.read()
            poems.write(text)
            poems.write("\n\n")
            file.close()

    poems.close()


if __name__ == "__main__":
    work()
```
